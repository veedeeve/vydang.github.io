# Independent Practice
**Objective:** Given Canada's adjusted net national income per capita, predict the net income in 2020

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
train_pract_df = pd.read_csv('canada_per_capita_income.csv')
train_pract_df.head(3)
```

```python
train_pract_df.info()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(train_pract_df.year, train_pract_df['per capita income (US$)'], color='red', marker='+')
```

```python
pract_reg = linear_model.LinearRegression()
pract_reg.fit(train_pract_df[['year']], train_pract_df[['per capita income (US$)']])
```

```python
pract_reg.predict([[2020]])
```
